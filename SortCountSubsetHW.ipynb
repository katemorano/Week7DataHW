```python
## find all incidents of defective pavements causing accidents 
## and sort by borough
defectsManhattan = crashes.query("(`CONTRIBUTING FACTOR VEHICLE 1` == 'Pavement Slippery') and (BOROUGH == 'MANHATTAN')")
defectsBrooklyn = crashes.query("(`CONTRIBUTING FACTOR VEHICLE 1` == 'Pavement Slippery') and (BOROUGH == 'BROOKLYN')")
defectsStatenIsland = crashes.query("(`CONTRIBUTING FACTOR VEHICLE 1` == 'Pavement Slippery') and (BOROUGH == 'STATEN ISLAND')")
defectsQueens = crashes.query("(`CONTRIBUTING FACTOR VEHICLE 1` == 'Pavement Slippery') and (BOROUGH == 'QUEENS')")
defectsBronx = crashes.query("(`CONTRIBUTING FACTOR VEHICLE 1` == 'Pavement Slippery') and (BOROUGH == 'BRONX')")
```
